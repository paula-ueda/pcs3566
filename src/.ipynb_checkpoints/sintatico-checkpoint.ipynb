```python
class Estado():
    def __init__(self, maquina, estado):
        self.maquina = maquina
        self.estado = estado
```

```python
class Pilha():
    def __init__(self):
        self.pilha = []
    def push(self, saida, entrada, nexttest, retorno):
        self.pilha.append((saida, nexttest, retorno))
    def pop(self):
        if pilha.pilha:
            elemento = self.pilha[-1]
            del(self.pilha[-1])
            return elemento
        else:
            return None
```

```python
def automato_program(lista_token, pilha, nexttest=None, estado=0):
    success = None
    if estado == 0:
        pilha.push(Estado("Program", 0), None, Estado("Program", 1))
        proximo_estado = Estado("Bstatement", 0)
    elif estado == 1:
        if nexttest is None:
            pilha.push(Estado("Program", 1), "Int", Estado("Program", 1))
            proximo_estado = Estado("Bstatement", 0)
        elif nexttest == "Int":
            pilha.push(Estado("Program", 1), False, Estado("Program", 2))
            proximo_estado = Estado("Int", 0)
    elif estado == 2:
        token = lista_token[0]
        del(lista_token[0])
        if token.nome == "END":
            pilha.push(Estado("Program", 2), None, Estado("Program", 3))
            proximo_estado = None
        else:
            proximo_estado = None
            success = False
    elif estado == 3:
        if len(lista_token) == 0:
            proximo_estado = None
            success = True
        else:
            success = False
            proximo_estado = None
            
    return success, pilha, proximo_estado
```

```python
def automato_bstatement(lista_token, pilha, nexttest=None, estado=0):
    success = None
    if estado == 0:
        pilha.push(Estado("Bstatement", 0), "Assign", Estado("Bstatement", 1))
        proximo_estado = Estado("Int", 0)
    elif estado == 1:
        if nexttest == "Assign":
            pilha.push(Estado("Bstatement", 1), "Read", Estado("Bstatement", 2))
            proximo_estado = Estado("Assign", 0)
        elif nexttest == "Read":
            pilha.push(Estado("Bstatement", 1), "Data", Estado("Bstatement", 2))
            proximo_estado = Estado("Read", 0)
        elif nexttest == "Data":
            pilha.push(Estado("Bstatement", 1), "Print", Estado("Bstatement", 2))
            proximo_estado = Estado("Data", 0)
        elif nexttest == "Print":
            pilha.push(Estado("Bstatement", 1), "Goto", Estado("Bstatement", 2))
            proximo_estado = Estado("Print", 0)
        elif nexttest == "Goto":
            pilha.push(Estado("Bstatement", 1), "If", Estado("Bstatement", 2))
            proximo_estado = Estado("Goto", 0)
        elif nexttest == "If":
            pilha.push(Estado("Bstatement", 1), "For", Estado("Bstatement", 2))
            proximo_estado = Estado("If", 0)
        elif nexttest == "For":
            pilha.push(Estado("Bstatement", 1), "Next", Estado("Bstatement", 2))
            proximo_estado = Estado("For", 0)
        elif nexttest == "Next":
            pilha.push(Estado("Bstatement", 1), "Dim", Estado("Bstatement", 2))
            proximo_estado = Estado("Next", 0)
        elif nexttest == "Dim":
            pilha.push(Estado("Bstatement", 1), "Def", Estado("Bstatement", 2))
            proximo_estado = Estado("Dim", 0)
        elif nexttest == "Def":
            pilha.push(Estado("Bstatement", 1), "Gosub", Estado("Bstatement", 2))
            proximo_estado = Estado("Def", 0)
        elif nexttest == "Gosub":
            pilha.push(Estado("Bstatement", 1), "Return", Estado("Bstatement", 2))
            proximo_estado = Estado("Gosub", 0)
        elif nexttest == "Return":
            pilha.push(Estado("Bstatement", 1), "Remark", Estado("Bstatement", 2))
            proximo_estado = Estado("Read", 0)
        elif nexttest == "Remark":
            pilha.push(Estado("Bstatement", 1), False, Estado("Bstatement", 2))
            proximo_estado = Estado("Remark", 0)

    elif estado == 2:
        success = True
        proximo_estado = None
        
    return success, pilha, proximo_estado
```

```python
def automato_assign(lista_token, pilha, nexttest=None, estado=0):
    success = None
    if estado == 0:
        token = lista_token[0]
        del(lista_token[0])
        if token.name == "LET":
            proximo_estado = Estado("Assign", 1)
        else:
            proximo_estado = None
            success = False
    elif estado == 1:
        pilha.push(Estado("Assign", 1), False, Estado("Assign", 2))
        proximo_estado = Estado("Var", 0)
    elif estado == 2:
        token = lista_token[0]
        del(lista_token[0])
        if token.name == "=":
            proximo_estado = Estado("Assign", 3)
        else:
            proximo_estado = None
            success = False
    elif estado == 3:
        pilha.push(Estado("Assign", 3), False, Estado("Assign", 4))
        proximo_estado = Estado("Exp", 0)
        
    elif estado == 4:
        success = True
        proximo_estado = None
        
    return success, pilha, proximo_estado
```
